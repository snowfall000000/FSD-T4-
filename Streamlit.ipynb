Example - 2 Sidebar,Columns,Expanders

```python
%%writefile 1_hello.py
import streamlit as st
st.set_page_config(page_title='hello Streamlit',page_icon='ðŸ˜Ž',layout='centered')
st.title("Faculty profile demo")
st.markdown("This example shows how to use **sidebar,columns,expanders**")

#Sidebar important for filters and settings
st.sidebar.header("profile Settings")

faculty_name=st.text_input("faculty name",'Tejas Thakkar')
department=st.sidebar.selectbox("Department",['CE','IT','CSE','AIML'])
experience=st.sidebar.slider("Years of Experience",0,40,10)

st.sidebar.markdown('---')
st.sidebar.write('you can put filers,toggles etc in sidebar')

col1,col2 = st.columns([1,2])

with col1:
    st.subheader("Basic info")
    st.write(f"**Name:** {faculty_name}")
    st.write(f"**Department:** {department}")
    st.write(f"**Experience:** {experience} years")
with col2:
    st.subheader("About")
    st.write("""Use this area to show detailed information about the faculty members""")
    
with st.expander("Show Courses handled"):
    st.write('python-1')
    st.write('python-2')
    st.write('Digital Electronics')
with st.expander("Show Publications"):
    st.write('1. Research paper A(2024)')
    st.write('2. Research paper A(2025)')
```

# Part - 2 Input Widgets & interactivity

Example - 3 Text Inputs

```python
%%writefile 3_text_inputs.py
import streamlit as st

st.title("Text Input Demo")

name = st.text_input("Enter Your Name : ")
comments = st.text_area("Any Comments or Feedback")

st.write("Live Output")

if name:
    st.write("Hello, **{name}**")
if comments:
    st.write('your Comments : ')
    st.write(comments)
```

# Number Inputes & Sliders

```python
%%writefile 4_number_slider.py
import streamlit as st

st.title('number Input & Slider Demo')

age = st.number_input('Enter your age : ',min_value=0,max_value=100,value=25)
rating = st.slider('rate this session(1-10)',min_value=1,max_value=10,value=5)

st.write(f"Your Age : {age}")
st.write(f"You rated this session : {rating}/10")
```

# Example - 5 Selection widgets

```python
%%writefile 5_selection_widget.py
import streamlit as st

course = st.selectbox('select Course : ',['python','FSD','DE','PS'])
preferred_days = st.multiselect('prefered Days for extra lectures',
                                ['Monday','Tuesday','Wednesdat','Thusday','Friday','Saturday'])
delievery_mode = st.radio('prefered Delievery Mode',
                         ['offline','online','hybrid'])

subscribe = st.checkbox("Subscribe to course updates")
st.write('---')
st.write(f"**Course : ** {course}")
st.write(f"**Prefered Days : ** {','.join(preferred_days)
if preferred_days else None}")
st.write(f"**Delievery Mode : ** {delievery_mode}")
st.write(f"**Subscribe:** {'Yes' if subscribe else 'No'}")
```
