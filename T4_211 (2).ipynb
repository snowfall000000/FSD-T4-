```python
def sum_numbers(*args):
    result=0
    for num in args:
        result+=num
    print("sum = ",result)
print("similer method overloading concept")
print("single argument : ",end="")
sum_numbers(10)
print("double argument : ",end="")
sum_numbers(10,2)
print("multiple argument : ",end="")
sum_numbers(1,2,3,4)
```

```python
from abc import ABC,abstractmethod
class shape(ABC):
    def __init__(self,shape_name):
        self.shape_name=shape_name
    @abstractmethod
    def draw(self):
        pass
class circle(shape):
    def __init__(self):
        super().__init__("cicrle")
    def draw(self):
        print("drawing :",self.shape_name)
obj=circle()
obj.draw()
```

```python
class book:
    def __init__(self,name,n,authors,publisher,ISBN,year):
        self.name=name
        self.n=n
        self.authors=authors
        self.publisher=publisher
        self.ISBN=ISBN
        self.year=year
    def display(self):
        print("book name : ",self.name)
        print("no of author : ",self.n)
        print("list of authors : ",self.authors)
        print("publisher : ",self.publisher)
        print("ISBN no : ",self.ISBN)
        print("year : ",self.year)
class Coursebook(book):
    def __init__(self,name,n,authors,publisher,ISBN,year,course):
        super().__init__(name,n,authors,publisher,ISBN,year)
        self.course=course
    def display(self):
        super().display()
        print("course name : ",self.course)
```

```python
obj1=Coursebook("harry potter",1,["mihir"],"Aryan",2468,2020,"movie")
obj1.display()
```

```python
class student:
    def __init__(self,rollno,name,age,totalnum):
        self.rollno=rollno
        self.name=name
        self.age=age
        self.totalnum=totalnum
    def display():
        print("students rollno : ",self.rollno)
        print("students name : ",self.name)
        print("students age :",self.age)
        print("total num : ",self.totalnum)
    def __equ__(self,other):
        if self.totalnum==other.totalnum:
            return True
        else:
            return False
```

```python
s1=student(1,"aryan",20,85)
s2=student(2,"mihir",19,89)
if s1==s2:
    print(f"{s1.name} and {s2.name} have same marks!!")
```
