```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
font={'color':'red'}
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,1,1)
plt.title("first half",fontdict=font)
plt.plot(x,y)
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,1,2)
plt.title("second half",fontdict=font)
plt.tight_layout()
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(20,10))
font={'color':'red','size':20}
font1={'color':'blue','size':20}

x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,3,1)
plt.title("first half",fontdict=font)
plt.xlabel("number",fontdict=font1,loc='left')
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,3,2)
plt.title("second half",fontdict=font)
plt.xlabel("number",fontdict=font1,loc='left')
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([50,10,80,40])
plt.subplot(2,3,3)
plt.title("third half",fontdict=font)
plt.xlabel("number",fontdict=font1,loc='left')
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([90,29,45,23])
plt.subplot(2,3,4)
plt.title("forth half",fontdict=font)
plt.xlabel("number",fontdict=font1,loc='left')
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([80,40,30,45])
plt.subplot(2,3,5)
plt.title("fifth half",fontdict=font)
plt.xlabel("number",fontdict=font1,loc='left')
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([4,7,9,1])
plt.subplot(2,3,6)
plt.title("sixth half",fontdict=font)
plt.xlabel("number",fontdict=font1,loc='left')
plt.plot(x,y)

plt.suptitle("my data",fontsize=50)
plt.tight_layout()
plt.show()
```

AREA CHART

```python
import matplotlib.pyplot as plt
import numpy as np
x=range(1,6)
y=[1,4,6,8,4]
plt.fill_between(x,y,color='blue')
plt.plot(x,y,color='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expanse=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expanse,color='red')
plt.fill_between(time,income,expanse,where=(income>expanse),color='green',alpha=0.25,label='positive')
plt.fill_between(time,income,expanse,where=(income<=expanse),color='red',alpha=0.25,label='negative')
plt.xlabel("time")
plt.ylabel("income/expanse comparision")
plt.title("Area fill between two lines")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,6))
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expanse=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expanse,color='red')
plt.fill_between(time,income,expanse,where=(income>expanse),color='green',alpha=0.25,label='positive',interpolate=True)
plt.fill_between(time,income,expanse,where=(income<=expanse),color='red',alpha=0.25,label='negative',interpolate=True)
plt.xlabel("time")
plt.ylabel("income/expanse comparision")
plt.title("Area fill between two lines")
plt.legend()
plt.show()
```
